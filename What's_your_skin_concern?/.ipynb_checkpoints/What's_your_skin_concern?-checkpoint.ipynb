### 고민 해결 제품 추천

- 고민 우선순위와 성분 등급에 따른 점수화를 통해 상위 5개 등급 추천

```python
df_Hydration = pd.read_csv('Hydration.csv')

for i in range(len(df_Hydration)):
    df_Hydration['Ingredients'][i] = df_Hydration['Ingredients'][i].lower()
    
df_Exfoliant = pd.read_csv('Exfoliant.csv')

for i in range(len(df_Exfoliant)):
    df_Exfoliant['Ingredients'][i] = df_Exfoliant['Ingredients'][i].lower()
```

```python
user_concern = ['Hydration', 'Exfoliant']
```

```python
df_product['Ingredients'][4].split(', ')
```

```python
count = []
score = 0
df_user = pd.DataFrame(columns = ['Product', 'Score'])

for i in range(133):
    count = []
    score = 0
    ing = df_product['Ingredients'][i].split(', ')
    
    for j in range(len(df_Hydration)):
        if df_Hydration['Ingredients'][j] in ing:
            count.append(5)
    
    count = []       
    for j in range(len(df_Exfoliant)):
        if df_Exfoliant['Ingredients'][j] in ing:
            count.append(4)
    
    for n in count:
            score += n

    df_user.loc[i] = {'Product':df_product['Product'][i], 'Score':score}
```

```python
df_user.sort_values(by='Score', ascending=False)
```
