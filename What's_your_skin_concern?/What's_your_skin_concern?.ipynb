### 고민 해결 제품 추천

- 고민 우선순위와 성분 등급에 따른 점수화를 통해 상위 5개 등급 추천

```python
user_concern = ['Hydration', 'Exfoliant']
priority = [10, 5]
```

```python
df_user = pd.DataFrame(columns = ['Product', 'Score'])

#고민별 성분데이터 지정
df_category1 = df_category[df_category['Category'] == user_concern[0]]
df_category1 = df_category1.reset_index(drop=True)
if len(user_concern) > 1:
    df_category2 = df_category[df_category['Category'] == user_concern[1]]
    df_category2 = df_category2.reset_index(drop=True)
if len(user_concern) > 2:
    df_category3 = df_category[df_category['Category'] == user_concern[2]]
    df_category3 = df_category3.reset_index(drop=True)
if len(user_concern) > 3:
    df_category4 = df_category[df_category['Category'] == user_concern[3]]
    df_category4 = df_category4.reset_index(drop=True)
if len(user_concern) > 4:
    df_category5 = df_category[df_category['Category'] == user_concern[4]]
    df_category5 = df_category5.reset_index(drop=True)

#점수 계산
for i in range(len(df_product)):
    score = 0
    ing_list = df_product['Ingredients'][i].split(', ')
    
    #고민 1
    for j in range(len(df_category1)):
        
        score1 = 0
        ingredient = df_category1['Ingredients'][j]
        rate = df_category1['Rate'][j]

        if ingredient in ing_list:
            if rate == 'BEST':
                score1 = priority[0]*5
            elif rate == 'GOOD':
                score1 = priority[0]*3
            elif rate == 'AVERAGE':
                score1 = priority[0]*2
            elif rate == 'POOR':
                score1 = priority[0]*1
                
        score += score1
            
    #고민 2
    try:
        for j in range(len(df_category2)):

            score2 = 0
            ingredient = df_category2['Ingredients'][j]
            rate = df_category2['Rate'][j]

            if ingredient in ing_list:
                if rate == 'BEST':
                    score2 = priority[1]*5
                elif rate == 'GOOD':
                    score2 = priority[1]*3
                elif rate == 'AVERAGE':
                    score2 = priority[1]*2
                elif rate == 'POOR':
                    score2 = priority[1]*1

            score += score2
    except:
        pass

    #고민 3
    try:
        for j in range(len(df_category3)):

            score3 = 0
            ingredient = df_category3['Ingredients'][j]
            rate = df_category3['Rate'][j]

            if ingredient in ing_list:
                if rate == 'BEST':
                    score3 = priority[2]*5
                elif rate == 'GOOD':
                    score3 = priority[2]*3
                elif rate == 'AVERAGE':
                    score3 = priority[2]*2
                elif rate == 'POOR':
                    score3 = priority[2]*1

            score += score3
    except:
        pass

    #고민 4
    try:
        for j in range(len(df_category4)):

            score4 = 0
            ingredient = df_category4['Ingredients'][j]
            rate = df_category4['Rate'][j]

            if ingredient in ing_list:
                if rate == 'BEST':
                    score4 = priority[3]*5
                elif rate == 'GOOD':
                    score4 = priority[3]*3
                elif rate == 'AVERAGE':
                    score4 = priority[3]*2
                elif rate == 'POOR':
                    score4 = priority[3]*1

            score += score4
    except:
        pass

    #고민 5
    try:
        for j in range(len(df_category5)):

            score5 = 0
            ingredient = df_category5['Ingredients'][j]
            rate = df_category5['Rate'][j]

            if ingredient in ing_list:
                if rate == 'BEST':
                    score5 = priority[4]*5
                elif rate == 'GOOD':
                    score5 = priority[4]*3
                elif rate == 'AVERAGE':
                    score5 = priority[4]*2
                elif rate == 'POOR':
                    score5 = priority[4]*1

            score += score4
    except:
        pass
    

    df_user.loc[i] = {'Product':df_product['Product'][i], 'Score':score}
```

```python
for i in range(len(df_product)):
    ing_list = df_product['Ingredients'][i].split(', ')
    similarity = []
    
    #효능 1
    concern1 = []        
    for j in range(len(df_category1)):
        ingredient = df_category1['Ingredients'][j]
        
        if ingredient in ing_list:
            concern1.append(ingredient)
            
        #유사도 측정
        while len(concern1) != len(ing_list):
            concern1.append('none')
            
        sim = jaccard_similarity_score(concern1, ing_list)
        similarity.append(sim)
        
    #효능 2
    concern2 = []        
    for j in range(len(df_category2)):
        ingredient = df_category2['Ingredients'][j]
        
        if ingredient in ing_list:
            concern2.append(ingredient)
            
        #유사도 측정
        while len(concern2) != len(ing_list):
            concern2.append('none')
            
        sim = jaccard_similarity_score(concern2, ing_list)
        similarity.append(sim)
    
    effect = []
    for n in range(len(similarity)): 
        if similarity[n] > 0:
            effect.append(concern[n])
    
    df_effect.loc[i] = {'Product':df_product['Product'][i], 'Effect':effect}
```
