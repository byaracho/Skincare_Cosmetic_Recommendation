```python
import codecs
 
infile = codecs.open('theordinary.csv', 'r', encoding='utf-8')
outfile = codecs.open('theordinary2.csv', 'w', encoding='euc_kr')
 
for line in infile:
    line = line.replace(u'\xa0', ' ')    # 가끔 \xa0 문자열로 인해 오류가 발생하므로 변환
    outfile.write(line)
    
infile.close()
outfile.close()
```
